**1. Data Exploration**

```python
import pandas as pd
import numpy as np #numerical operation
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/content/Lesson_4_student_grades_dataset.csv', encoding = 'latin-1')
print(data.head())
```

The potential impact of each feature hours of study, attendance percentage, previous exam scores, extracurricular activities, travel time to school depends on the previous grades of students. The data will be utilize in predicting grades depending on the performance of student in school.

**2 Data Processing**

```python
print("Total number of missing data per features:")
print(data.isnull().sum())
```

Normalizing the numerical features is important for better result and improvement of the student grade in the future.

**3. Train-Test Split:**

```python
x = data.drop('Grades', axis = 1)
y = data ['Grades']

x_train, x_test, y_train, y_test, = train_test_split(x,y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
```

**4. Linear Regression Model and 5. Model Trainig**

**Hours of Study**

```python
h = data[['Hours_of_Study']]
g = data['Grades']


model = LinearRegression()
model.fit(h,g)
```

```python
plt.scatter(data['Hours_of_Study'], data['Grades'])
plt.plot(h, model.predict(h), color='red')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('Hours of Study')

```
